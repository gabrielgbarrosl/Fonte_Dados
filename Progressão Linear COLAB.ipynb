```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score

# Dados
data = {
    "Experience Years": [1.1,1.2,1.3,1.5,2,2.2,2.5,2.9,3,3.2,3.2,3.5,3.7,3.8,3.9,4,4,4.1,4.3,4.5,4.7,4.9,5.1,5.3,5.5,5.9,6,6.2,6.5,6.8,7.1,7.9,8.2,8.5,8.7,9,9.5,9.6,10.3,10.5],
    "Salary": [39343,42774,46205,37731,43525,39891,48266,56642,60150,54445,64445,60000,57189,60200,63218,55794,56957,57081,59095,61111,64500,67938,66029,83088,82200,81363,93940,91000,90000,91738,98273,101302,113812,111620,109431,105582,116969,112635,122391,121872]
}

df = pd.DataFrame(data)

# Modelo
X = df[["Experience Years"]]
y = df["Salary"]

model = LinearRegression()
model.fit(X, y)

# Previsões
y_pred = model.predict(X)

# Resultados
print("Coeficiente (slope):", model.coef_[0])
print("Intercept:", model.intercept_)
print("R²:", r2_score(y, y_pred))

# Plot
plt.scatter(X, y, color='red') # Alterado aqui para mudar a cor para vermelho
plt.plot(X, y_pred, color='blue') # Alterado aqui para mudar a cor para azul
plt.xlabel("Experience Years")
plt.ylabel("Salary")
plt.title("Regressão Linear de Salário vs Experiência")
plt.show()
```
